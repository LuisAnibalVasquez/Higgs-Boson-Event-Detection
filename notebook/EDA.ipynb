```python
aux = pd.DataFrame(df.drop(columns=['EventId','Label'], axis=1), columns=df.drop(columns=['EventId','Label'], axis=1).columns)
skewed_data = aux.skew().sort_values(ascending=False)
skewed_data
```

```python
aux = pd.DataFrame(df.drop(columns=['EventId','Label'], axis=1), columns=df.drop(columns=['EventId','Label'], axis=1).columns)

for c in aux.columns:
    sm.qqplot(aux[c], line ='45')
    plt.show()
```
